```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# =========================
# 1. Cek file dulu (biar tidak error)
# =========================
path = '/content/drive/MyDrive/Colab Notebooks/Computer Vision/dataset/lena.jpg'

if not os.path.exists(path):
    print("❌ File tidak ditemukan di path:", path)
else:
    # =========================
    # 2. Load gambar
    # =========================
    img = cv2.imread(path)

    if img is None:
        print("❌ Gambar gagal dibaca!")
    else:
        # =========================
        # 3. Konversi grayscale
        # =========================
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # =========================
        # 4. Tampilkan RGB & Grayscale
        # =========================
        plt.figure(figsize=(10,4))

        plt.subplot(1,2,1)
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.title("RGB")
        plt.axis('off')

        plt.subplot(1,2,2)
        plt.imshow(gray, cmap='gray')
        plt.title("Grayscale")
        plt.axis('off')

        plt.show()

        # =========================
        # 5. Ukuran gambar
        # =========================
        h, w = img.shape[:2]
        print("Height = {}, Width = {}".format(h, w))

        # =========================
        # 6. Nilai pixel
        # =========================
        (B, G, R) = img[100, 100]
        print("R = {}, G = {}, B = {}".format(R, G, B))

        # =========================
        # 7. Histogram RGB
        # =========================
        plt.figure(figsize=(10,4))

        colors = ('b', 'g', 'r')
        for i, c in enumerate(colors):
            hist = cv2.calcHist([img], [i], None, [256], [0,256])
            plt.plot(hist, color=c, label=f'Channel {c}')

        plt.title("Histogram RGB")
        plt.legend()
        plt.tight_layout()
        plt.show()

        # =========================
        # 8. Histogram Grayscale
        # =========================
        plt.figure(figsize=(6,4))
        plt.hist(gray.ravel(), bins=256, color='black')
        plt.title("Histogram Grayscale")
        plt.tight_layout()
        plt.show()
```

📘 Analisis Citra Digital: RGB, Grayscale, dan Histogram
#1. Pembacaan Citra

Citra dibaca menggunakan fungsi cv2.imread(), yang mengubah file gambar menjadi array (matriks piksel).

img = cv2.imread(path)
Jika gambar berhasil dibaca → berbentuk array NumPy
Jika gagal → bernilai None
Penyebab umum gagal: path salah atau file tidak tersedia

#2. Konversi ke Grayscale

Citra RGB diubah menjadi grayscale menggunakan:

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
RGB memiliki 3 channel (Red, Green, Blue)
Grayscale hanya 1 channel (intensitas)
Tujuan:
Menyederhanakan data
Mempercepat komputasi
Digunakan pada banyak proses (deteksi tepi, segmentasi)

#3. Visualisasi Citra

Untuk menampilkan citra digunakan matplotlib:

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
OpenCV menggunakan format BGR
Matplotlib menggunakan RGB
Harus dikonversi agar warna tidak terbalik

Untuk grayscale:

plt.imshow(gray, cmap='gray')

#4. Dimensi Citra

Ukuran citra diperoleh dari:

h, w = img.shape[:2]
h = tinggi (height)
w = lebar (width)

#5. Nilai Piksel

Nilai piksel dapat diakses dengan:

(B, G, R) = img[100, 100]
Setiap channel memiliki nilai 0–255
0 = gelap
255 = terang

#6. Histogram RGB

Histogram menunjukkan distribusi intensitas warna:

cv2.calcHist([img], [i], None, [256], [0,256])

Keterangan:

[img] → input gambar
[i] → channel (0=B, 1=G, 2=R)
[256] → jumlah bin
[0,256] → rentang intensitas

Fungsi histogram:

Mengetahui distribusi warna
Melihat dominasi warna tertentu

#7. Histogram Grayscale

Histogram grayscale dibuat dengan:

plt.hist(gray.ravel(), bins=256)
ravel() mengubah array menjadi 1 dimensi
Menampilkan distribusi intensitas seluruh piksel

#8. Interpretasi Histogram

Histogram dapat digunakan untuk analisis:

a. Kontras
Menyebar luas → kontras tinggi
Menumpuk → kontras rendah
b. Kecerahan
Dominan kiri → gelap
Dominan kanan → terang
c. Warna Dominan
Channel tertentu tinggi → warna tersebut dominan

#9. Tujuan Praktikum
Memahami representasi citra sebagai matriks
Mengetahui perbedaan RGB dan grayscale
Menganalisis citra menggunakan histogram
Menjadi dasar untuk pengolahan citra lanjutan

#🧠 Kesimpulan

Citra digital direpresentasikan sebagai matriks piksel dengan nilai intensitas tertentu. Citra RGB memiliki tiga channel warna, sedangkan grayscale hanya satu channel. Histogram digunakan untuk menganalisis distribusi intensitas piksel sehingga dapat diketahui kontras, kecerahan, dan karakteristik citra. Pemahaman ini menjadi dasar dalam berbagai teknik pengolahan citra seperti enhancement, segmentasi, dan deteksi objek.
